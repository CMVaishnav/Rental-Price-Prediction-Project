```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
data = pd.read_csv("/Users/cm/Downloads/Indian_housing_Delhi_data.csv")

# Convert relevant columns for visualization
data['house_size_numeric'] = data['house_size'].str.extract('(\d+)').astype(float)
data_cleaned = data.dropna(subset=['price', 'location'])

# Set a consistent style
sns.set_theme(style="whitegrid")



# Visualization 1: Rental Prices by Location (Top 10 Locations)
top_locations = data_cleaned['location'].value_counts().head(10).index
filtered_data = data_cleaned[data_cleaned['location'].isin(top_locations)]
plt.figure(figsize=(12, 8))
sns.boxplot(data=filtered_data, x="location", y="price", palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Rental Prices by Location (Top 10 Locations)")
plt.xlabel("Location")
plt.ylabel("Rental Price (INR)")
plt.tight_layout()
plt.show()

```
